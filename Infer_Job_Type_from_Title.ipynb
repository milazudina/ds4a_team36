```python
# Load required libraries

import pandas as pd
import numpy as np
```

```python
# Code adapted from Hannah's EDA

filename = './Indeed_Elroy_2018/indeed_job_dataset.csv'
df = pd.read_csv(filename, index_col = None, header = 0)
df = df[df['Skill'].notna()]
df.drop(columns = ['python', 'sql', 'machine learning', 'r', 'hadoop', 'tableau', 'sas', 'spark', 'java','Others'], inplace = True)
df.drop(columns = ['Consulting and Business Services','Internet and Software', 'Banks and Financial Services', 'Health Care', 'Insurance', 'Other_industries'], inplace = True)
df.drop(columns = ['CA', 'NY', 'VA', 'TX', 'MA', 'IL', 'WA', 'MD', 'DC', 'NC', 'Other_states'], inplace = True)
df.insert(4,'Inferred_Job_Type','')
df.insert(6,'Inferred_vs_Data','')
df.head(5)
```

```python
# these 3 rules can achieve accuracy of 93.76% 

#df.loc[df['Job_Title'].str.contains('data scientist', case = False) , 'Inferred_Job_Type'] = 'data_scientist'
#df.loc[df['Job_Title'].str.contains('data analyst', case = False) , 'Inferred_Job_Type'] = 'data_analyst'
#df.loc[df['Job_Title'].str.contains('data engineer', case = False) , 'Inferred_Job_Type'] = 'data_engineer'
```

```python
# these 3 rules can achieve accuracy of 96.35% 

df.loc[df['Job_Title'].str.contains('analyst', case = False) , 'Inferred_Job_Type'] = 'data_analyst'
df.loc[df['Job_Title'].str.contains('scientist', case = False) , 'Inferred_Job_Type'] = 'data_scientist'
df.loc[df['Job_Title'].str.contains('engineer', case = False) , 'Inferred_Job_Type'] = 'data_engineer'
```

```python
df['Inferred_vs_Data'] = np.where(df['Inferred_Job_Type'] == df['Job_Type'], 'same', 'different')
df.head(10)
```

```python
df.groupby(['Inferred_vs_Data']).size().sort_values(ascending=False)
```
